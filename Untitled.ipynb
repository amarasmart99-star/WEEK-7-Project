```python
import requests

API_KEY =" YOUR_API_KEY_HERE"

CITIES = ["Kano", "Lagos", "Abuja", "London", "New York"]

url = "https://api.openweathermap.org/data/2.5/weather"

raw_records = []

for city in CITIES:
    params = {"q": city, "appid": API_KEY, "units": "metric"}
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        raw_records.append(response.json())
        print(f"OK - {city}")
    else:
        print(f"FAILED - {city} (status {response.status_code})")

print(f"\nTotal cities collected: {len(raw_records)}")
```

```python
import pandas as pd

rows = []

for r in raw_records:
    rows.append({
        "city": r["name"],
        "country": r["sys"]["country"],
        "temperature_c": r["main"]["temp"],
        "feels_like_c": r["main"]["feels_like"],
        "humidity_pct": r["main"]["humidity"],
        "weather_condition": r["weather"][0]["main"],
        "weather_description": r["weather"][0]["description"],
        "wind_speed_mps": r["wind"]["speed"],
        "recorded_at_utc": r["dt"]
    })

df = pd.DataFrame(rows)

# Fix data types
df["temperature_c"] = df["temperature_c"].astype(float)
df["humidity_pct"] = df["humidity_pct"].astype(int)
df["recorded_at_utc"] = pd.to_datetime(df["recorded_at_utc"], unit="s")

df
```

```python
# Save as CSV
df.to_csv("weather_data.csv", index=False)
print("Saved weather_data.csv")

# Save as SQLite database
import sqlite3

conn = sqlite3.connect("weather_data.db")
df.to_sql("weather", conn, if_exists="replace", index=False)
conn.close()
print("Saved weather_data.db")
```

```python
print("--- Basic Weather Analysis ---\n")

hottest = df.loc[df["temperature_c"].idxmax()]
coldest = df.loc[df["temperature_c"].idxmin()]
most_humid = df.loc[df["humidity_pct"].idxmax()]

print(f"Hottest city: {hottest['city']} ({hottest['temperature_c']}°C)")
print(f"Coldest city: {coldest['city']} ({coldest['temperature_c']}°C)")
print(f"Most humid city: {most_humid['city']} ({most_humid['humidity_pct']}%)")

print("\nWeather conditions across cities:")
print(df[["city", "weather_condition", "weather_description"]].to_string(index=False))
```
